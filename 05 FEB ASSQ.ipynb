# Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

```python
'''In object-oriented programming (OOP), a class is a blueprint for creating objects. 
It defines the data and methods that an object will have. An object is an instance of a class. 
It is a real-world entity that has been created from a class.

For example, a class called Car could define the data and methods that a car has, such as its make, model, and year. 
An object of the Car class would be a specific car, such as a Toyota Camry or a Honda Accord.

The main difference between a class and an object is that a class is a blueprint, while an object is a real-world entity.
A class is abstract, while an object is concrete.

'''
```

# Q2. Name the four pillars of OOPs.

```python
'''The four pillars of OOPs are:

Abstraction: Abstraction allows programmers to focus on the essential features of an object without having to worry about how it works.
For example, a Car object might have an engine property, but the programmer does not need to know how the engine works in order to use the Car object.

Encapsulation: Encapsulation allows programmers to group together data and methods into a single unit. 
This makes the code easier to understand and maintain. 
For example, a Car object might have a start() method and a stop() method. 
These methods are encapsulated within the Car object, so the programmer does not need to know how they work in order to use them.

Inheritance: Inheritance allows a new class to inherit the properties and methods of an existing class. 
This allows programmers to reuse code and create more complex objects. 
For example, a SportsCar class might inherit from the Car class. 
This means that the SportsCar class would have all of the properties and methods of the Car class, 
plus any additional properties and methods that are specific to SportsCar objects.

Polymorphism: Polymorphism is the ability of an object to take on multiple forms. 
This allows programmers to write code that is more generic and flexible. 
For example, a Car object might be able to drive, but it might also be able to fly. 
This is because the Car object is polymorphic. It can take on the form of a car or a flying car.'''
```

# Q3. Explain why the __init__() function is used. Give a suitable example.

```python
'''The __init__() function is used to initialize the attributes of an object when it is created. 
It is a special method in Python that is called automatically when a new object is created from a class.

The __init__() function is often used to set the initial values of the object's attributes. 
For example, the following code defines a Car class with an __init__() function that sets the initial values of the make and model attributes:

class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model
        
        When a new Car object is created, the __init__() function is called and the make and model attributes are set to the values that were passed to the __init__() function.
        For example, the following code creates a new Car object with the make attribute set to "Toyota" and the model attribute set to "Camry":
        
        my_car = Car("Toyota", "Camry")
        
        '''
```

# Q4. Why self is used in OOPs?

```python
'''The self keyword is used to access the attributes and methods of the current object. 
For example, the following code defines a Car class with a get_make() method:

class Car:
    def __init__(self, make):
        self.make = make

    def get_make(self):
        return self.make
        
The get_make() method uses the self keyword to access the make attribute of the current object. 
For example, the following code creates a new Car object and calls the get_make() method:

my_car = Car("KIA")
print(my_car.get_make())

'''
```

#    # Q5. What is inheritance? Give an example for each type of inheritance.

```python
'''Inheritance is a fundamental concept in object-oriented programming where a class can inherit 
attributes and behaviors (methods) from another class. The class that inherits is called the "derived class" or "subclass," 
and the class being inherited from is called the "base class" or "superclass." 

Inheritance promotes code reuse and allows for the creation of hierarchical relationships between classes.

Here are some examples of each type of inheritance:

Single inheritance: The Dog class inherits from the Mammal class. 
This means that the Dog class can access all of the properties and methods of the Mammal class, such as the eat() method and the sleep() method.

class Mammal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print("I am eating.")

    def sleep(self):
        print("I am sleeping.")

class Dog(Mammal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def bark(self):
        print("Woof!")

my_dog = Dog("Rover", "Golden Retriever")
my_dog.bark()
my_dog.eat()
my_dog.sleep()

Multilevel inheritance: The Golden Retrieve class inherits from both the Dog class and the Mammal class. 
This means that the Golden Retrieve class can access all of the properties and methods of the Dog class and the Mammal class.

class Mammal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print("I am eating.")

    def sleep(self):
        print("I am sleeping.")

class Dog(Mammal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def bark(self):
        print("Woof!")

class GoldenRetriever(Dog):
    def __init__(self, name, breed, color):
        super().__init__(name, breed)
        self.color = color

    def fetch(self):
        print("I am fetching the ball.")

my_dog = GoldenRetriever("Rover", "Golden Retriever", "Yellow")
my_dog.bark()
my_dog.eat()
my_dog.sleep()
my_dog.fetch()

Hierarchical inheritance: The Mammal class is the parent class to both the Dog class and the Cat class. 
This means that the Dog class and the Cat class can access all of the properties and methods of the Mammal class.

class Mammal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print("I am eating.")

    def sleep(self):
        print("I am sleeping.")

class Dog(Mammal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def bark(self):
        print("Woof!")

class Cat(Mammal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def meow(self):
        print("Meow!")

my_dog = Dog("Rover", "Golden Retriever")
my_dog.bark()
my_dog.eat()
my_dog.sleep()

my_cat = Cat("Mittens", "Persian")
my_cat.meow()
my_cat.eat()
my_cat.sleep()

'''
```
